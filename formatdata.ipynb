## **Formats the data by creating new .csv files that combine the respective data from all seasons 1981-2024 in each: "basic stats", "advanced stats", and "mvp" folders**

```python
import pandas as pd
import numpy as np
import os
import re
```

```python
# Sets path for csv folder that contains folders of stats & mvp
csvfolder = os.getcwd() + "/csv"
# Gets list of subfolders of csv folder, removing .ds_store if in list
datafolders = os.listdir(csvfolder)
if '.DS_Store' in datafolders:
    datafolders.remove('.DS_Store')


# Goes through each stats and mvp folder in folder "csv" 
for folder in datafolders:
    folderpath = os.path.join(csvfolder, folder)
    csvfiles = os.listdir(folderpath)
    # Sorts to order seasons for easier understanding of df, removes .ds_store if in list
    csvfiles = sorted(csvfiles)
    if '.DS_Store' in csvfiles:
        csvfiles.remove('.DS_Store')
    combinedseasons = pd.DataFrame()
    
    # Goes through each csv file in the folder, creating a dataframe, adding season column, then concatenating with combinedseasons
    for csv in csvfiles:
        csvpath = os.path.join(folderpath, csv)
        if os.path.isfile(csvpath):
            df = pd.read_csv(csvpath)
            # Finds season from file name using regex, then creates season column with getseason string
            getseason = re.findall("[0-9]+-[0-9]+", csv)[0]
            df["Season"] = getseason
            # Concatenates the existing combinedseasons dataframe and the dataframe of the current csv
            combinedseasons = pd.concat([combinedseasons, df], ignore_index=True)
            
    # Saves a csv for each folder of all seasons data concatenated into a single dataframe (uncomment below to save)
    # combinedseasons.to_csv(folderpath + f'/combinedseasons{folder}.csv')

#example of last saved combinedseasons
combinedseasons
            
```
